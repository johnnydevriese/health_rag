```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Data based on PGA Tour official stats
years = list(range(1980, 2024))
average_distance = [
    256.89, 259.66, 256.89, 258.65, 259.61, 260.18, 261.58, 262.50, 263.50, 261.81, # 1980-1989
    262.75, 261.44, 260.52, 260.36, 261.84, 263.55, 266.49, 267.67, 270.63, 272.45, # 1990-1999
    273.18, 279.35, 279.84, 286.30, 287.32, 288.88, 289.35, 289.08, 287.74, 288.07, # 2000-2009
    287.49, 291.14, 290.07, 288.00, 289.85, 290.21, 291.06, 292.79, 295.29, 293.90, # 2010-2019
    296.40, 295.30, 299.80, 299.90  # 2020-2023
]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(years, average_distance, marker='o', linestyle='-')

# Add titles and labels
plt.title('PGA Tour Average Driving Distance (1980-2023)')
plt.xlabel('Year')
plt.ylabel('Average Driving Distance (Yards)')

# Customize the grid and ticks
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(range(1980, 2024, 5), rotation=45) # Show ticks every 5 years
plt.yticks(range(250, 310, 5)) # Set y-axis ticks every 5 yards
plt.ylim(250, 305) # Set y-axis limits

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# Data for Lowest Adjusted Scoring Average on PGA Tour (Byron Nelson Award winner)
# Source: Golf Monthly / PGA Tour Records
years = list(range(1980, 2024))
lowest_avg_score = [
    69.73, 69.80, 70.21, 70.61, 70.56, 70.36, 70.08, 70.09, 69.38, 69.49, # 1980-1989
    69.10, 69.59, 69.38, 68.90, 68.81, 69.06, 69.32, 68.98, 69.13, 68.43, # 1990-1999
    67.79, 68.81, 68.56, 68.41, 68.84, 68.66, 68.11, 67.79, 69.12, 68.05, # 2000-2009
    69.61, 68.86, 68.87, 68.95, 68.83, 68.94, 69.17, 68.85, 68.70, 69.06, # 2010-2019
    68.98, 69.30, 68.67, 68.63 # 2020-2023
]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(years, lowest_avg_score, marker='o', linestyle='-')

# Add titles and labels
plt.title('PGA Tour Lowest Adjusted Scoring Average per Year (Byron Nelson Award)')
plt.xlabel('Year')
plt.ylabel('Lowest Adjusted Scoring Average')

# Customize the grid and ticks
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(range(1980, 2024, 5), rotation=45) # Show ticks every 5 years
# Adjust y-axis for better visibility of the scoring range
plt.ylim(67.5, 71.0)
plt.gca().yaxis.set_major_locator(mticker.MultipleLocator(0.5)) # Ticks every 0.5 strokes

# Improve layout
plt.tight_layout()

# Show the plot
plt.show()
```
